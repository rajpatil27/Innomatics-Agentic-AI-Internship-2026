## Problem 1: Unique Words in a Sentence

## 🔹 Problem Statement

Given a sentence, identify all unique words and display:

1. Count of unique words  
2. The unique words themselves  

**Real-Time Applications:**
- Text analytics  
- NLP preprocessing  
- Search engine indexing  
- Chat analysis systems

## 🔹 Logic Explanation

1. Take the input sentence.
2. Convert the sentence into a list of words using split().
3. Convert the list into a set to remove duplicate words.
4. Count the number of unique words using len().
5. Print the unique words and their count.

```python
# Input sentence
sentence = "python is easy and python is powerful"

# Split sentence into words
words = sentence.split()

# Convert to set to get unique words
unique_words = set(words)

# Display results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

## Problem 2: Highest Salary from Employee Data

## 🔹 Problem Statement

Store employee names and their salaries and determine which employee has the highest salary.

**Real-Time Applications:**
- HR management systems
- Payroll analysis

## 🔹 Logic Explanation

1. Store employee data in a dictionary:
   - Key → Employee Name
   - Value → Salary

2. Use max() to find the highest salary.

3. Retrieve the employee name using the salary value.

4. Print the result.

```python
# Employee salary data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find highest salary
highest_salary = max(employees.values())

# Find employee with highest salary
for name, salary in employees.items():
    if salary == highest_salary:
        highest_paid_employee = name

print("Highest Salary:", highest_paid_employee, "-", highest_salary)
```

## Problem 3: Find Maximum and Minimum Values

## 🔹 Problem Statement

Find the maximum and minimum values from a list of numbers.

**Real-Time Applications:**
- Sales data analysis  
- Temperature monitoring systems  
- Performance metric evaluation

## 🔹 Logic Explanation

1. Initialize both max and min with the first element of the list.
2. Loop through the list.
3. Compare each element:
   - If greater than max → update max
   - If smaller than min → update min
4. Print the final maximum and minimum values.

```python
numbers = [45, 22, 89, 10, 66]

# Initialize
max_val = numbers[0]
min_val = numbers[0]

# Loop to find max and min
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("List:", numbers)
print("Max :", max_val)
print("Min :", min_val)
```

## Problem 4: Count Products Above a Price Threshold

## 🔹 Problem Statement

Count how many products have a price greater than 1000.

**Real-Time Applications:**
- E-commerce product filtering  
- Budget analysis

## 🔹 Logic Explanation

1. Store product prices in a list.
2. Initialize a counter to 0.
3. Loop through each price in the list.
4. If price > 1000 → increment the counter.
5. Print the final count.

```python
prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

## Problem 5: Calculate Attendance Percentage

## 🔹 Problem Statement

Calculate the attendance percentage of a student.

### Formula:
Attendance % = (Classes Attended / Total Classes) × 100

**Real-Time Applications:**
- Student performance tracking  
- Eligibility calculation for exams

## 🔹 Logic Explanation

1. Store total number of classes.
2. Store number of classes attended.
3. Use the attendance formula:
   
   attendance_percentage = (attended / total) * 100

4. Display the attendance percentage.

```python
total_classes = 50
attended_classes = 42

attendance_percentage = (attended_classes / total_classes) * 100

print(f"Attendance Percentage: {attendance_percentage:.2f}%")
```
