## Problem 2: Search Query Keyword Analysis

### Logic
1. Convert sentence to lowercase.
2. Remove punctuation.
3. Count word frequency.
4. Display words with frequency > 1.

```python
import string

query = "Buy mobile phone buy phone online"

query = query.lower().translate(str.maketrans('', '', string.punctuation))
words = query.split()

word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

result = {k: v for k, v in word_count.items() if v > 1}

print(result)
```

## Problem 3: Sensor Data Validation

## Problem Statement

A system records hourly sensor readings.  
Valid readings are those that are divisible by 2.

Identify all valid readings along with their corresponding hour index.

Real-Time Applications:
- IoT sensor monitoring
- Industrial equipment health tracking
- Environmental data validation

## Logic Explanation

1. Store all sensor readings in a list.
2. Use enumerate() to track both index (hour) and value.
3. Check if the reading is divisible by 2.
4. If valid, store the (hour, reading) as a tuple.
5. Display all valid sensor readings.

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## Problem 4: Email Domain Usage Analysis

## Problem Statement

You are given a list of email addresses.

Your task is to:
1. Extract the domain from each email.
2. Count how many users belong to each domain.
3. Calculate the percentage share of each domain.

Real-Time Applications:
- Customer segmentation
- Marketing analytics
- Organization email usage reports

## Logic Explanation

1. Store all email addresses in a list.
2. Extract the domain using split('@').
3. Use a dictionary to count domain occurrences.
4. Find the total number of emails.
5. Convert counts into percentages.
6. Display the domain-wise percentage usage.

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total = len(emails)

for domain, count in domain_count.items():
    print(f"{domain} : {(count/total)*100:.0f}%")
```

## Problem 5: Sales Spike Detection

## Problem Statement

A company records daily sales figures.

Your task is to:
1. Calculate the average sales.
2. Identify the days where sales are greater than 130% of the average.
3. Display those days along with the sales amount.

Real-Time Applications:
- Business performance monitoring
- Revenue anomaly detection
- Demand surge analysis

## Logic Explanation

1. Store daily sales in a list.
2. Calculate the average sales.
3. Compute the spike threshold as 1.3 × average.
4. Loop through the list using enumerate() to track day numbers.
5. If a day's sales exceed the threshold, mark it as a spike.
6. Display the spike days with sales values.

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales) / len(sales)
threshold = average_sales * 1.3

print("Average Sales:", average_sales)
print("Spike Threshold:", threshold)
print("Sales Spike Days:")

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print("Day", day, ":", value)
```

## Problem 6: Duplicate User ID Detection

## Problem Statement

A platform stores user IDs for login activity.

Your task is to:
1. Identify duplicate user IDs.
2. Count how many times each duplicate appears.
3. Display only the repeated user IDs.

Real-Time Applications:
- Detecting multiple logins
- Identifying fraudulent accounts
- Database cleanup

## Logic Explanation

1. Store all user IDs in a list.
2. Create a dictionary to count occurrences of each ID.
3. Loop through the list and update the count.
4. Filter IDs whose count is greater than 1.
5. Display duplicate IDs with their frequency.

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

for user in user_ids:
    if user in id_count:
        id_count[user] += 1
    else:
        id_count[user] = 1

print("Duplicate User IDs:")

for user, count in id_count.items():
    if count > 1:
        print(user, "→", count, "times")
```
