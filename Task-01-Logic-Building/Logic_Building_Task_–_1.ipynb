## Problem Statement

Given a username and password, check whether login is successful.

**Stored Credentials:**
- Username = "admin"
- Password = "1234"

### Requirements:
- Print "Login Successful" if both username and password match.
- Otherwise print "Invalid Credentials".

**Real-World Application:** Authentication Systems

## Logic Explanation

1. Store the correct username and password.
2. Take user input (or predefined values).
3. Compare entered credentials with stored credentials.

```python
# Task 1: User Login Check

# Step 1: Store correct credentials
correct_username = "admin"
correct_password = "1234"

# Step 2: Given credentials (can replace with input() if needed)
username = "admin"
password = "1234"

# Step 3: Check login condition
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

# Logic Building Task – 1  
## Task 2: Pass / Fail Analyzer  
### Data Science Internship – February 2026

## Problem Statement

Given the marks of students:

marks = [45, 78, 90, 33, 60]

### Requirements:
- A student passes if marks ≥ 50
- Count total pass students
- Count total fail students
- Print the final result clearly

**Real-World Application:** Academic Evaluation Systems

## Logic Explanation

1. Initialize two counters:
   - pass_count = 0
   - fail_count = 0

2. Loop through each mark in the list.

3. If mark ≥ 50:
      Increment pass_count
   Else:
      Increment fail_count

4. Print total students, pass count, and fail count.

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
        print(f"Marks: {mark} → PASS")
    else:
        fail_count += 1
        print(f"Marks: {mark} → FAIL")

print("\n--- Final Summary ---")
print("Total Students :", len(marks))
print("Total Passed   :", pass_count)
print("Total Failed   :", fail_count)
```

# Logic Building Task – 1  
## Task 3: Simple Data Cleaner  
### Data Science Internship – February 2026

## Problem Statement

Given a list of names:

names = [" Alice ", "bob", " CHARLIE "]

### Requirements:
- Remove extra spaces
- Convert all names to lowercase
- Print the cleaned list

**Real-World Application:** Data Preprocessing before Analysis

## Logic Explanation

1. Create an empty list to store cleaned names.
2. Loop through each name in the list.
3. Use:
   - strip() → to remove leading and trailing spaces
   - lower() → to convert text into lowercase
4. Append cleaned names into a new list.
5. Print original and cleaned lists.

```python
# Task 3: Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Original Names :", names)
print("Cleaned Names  :", cleaned_names)
```

## Task 4: Message Length Analyzer  
### Data Science Internship – February 2026

## Problem Statement

Given a list of messages:

messages = ["Hi", "Welcome to the platform", "OK"]

### Requirements:
- Print the length of each message
- Flag messages longer than 10 characters

**Real-World Application:** Text Filtering and Validation Systems

## Logic Explanation

1. Loop through each message in the list.
2. Use len() to calculate the length of the message.
3. Print the message and its length.
4. If length > 10:
      Flag the message as long.

```python
# Task 4: Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)

    print("Message :", message)
    print("Length  :", length)

    if length > 10:
        print("Flag    : Message is longer than 10 characters")

    print("-" * 40)
```

# Logic Building Task – 1  
## Task 5: Error Message Detector  
Data Science Internship – February 2026

## Problem Statement

Given system logs:

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

### Requirements:
- Count the number of "ERROR" entries
- Print the total error count

**Real-World Application:** Monitoring and Log Analysis Systems

## Logic Explanation

1. Initialize a counter variable (error_count = 0).
2. Loop through each log entry.
3. If log == "ERROR":
      Increment error_count.
4. Print total error count.

```python
# Task 5: Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
