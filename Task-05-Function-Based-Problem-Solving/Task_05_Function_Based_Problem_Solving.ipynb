## Problem 1: Social Media – Post Engagement Analyzer

## Problem Statement

Create a function to analyze engagement on a social media post.

Rules:
- Calculate total likes
- If total likes ≥ 1000 → Viral Post
- Otherwise → Normal Engagement

Real-Time Applications:
- Social media analytics platforms
- Marketing performance dashboards

## Logic Explanation

1. Create a function that accepts a list of likes.
2. Use a loop to calculate the total likes.
3. Check the total:
   - If ≥ 1000 → Viral Post
   - Else → Normal Engagement
4. Display total likes and post status.

```python
def analyze_post_engagement(likes_list):
    total_likes = 0

    for likes in likes_list:
        total_likes += likes

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


likes_data = [120, 340, 560, 90]

analyze_post_engagement(likes_data)
```

## Problem 2: Healthcare – Medicine Stock Alert System

## Problem Statement

Create a function to check whether the medicine stock is low or sufficient.

Rules:
- If stock is less than or equal to the minimum required level → "Low Stock – Reorder Needed"
- Otherwise → "Stock is Sufficient"

Return the stock status.

Real-Time Applications:
- Hospital inventory management
- Pharmacy stock monitoring
- Automated reorder systems

## Logic Explanation

1. Create a function that accepts current stock and minimum required stock.
2. Compare the current stock with the minimum required level.
3. If current stock is less than or equal to the minimum level → return low stock alert.
4. Otherwise → return sufficient stock message.
5. Display the result.

## Problem 3: Agriculture – Rainfall Adequacy Checker

## Problem Statement

Create a function to check whether the rainfall received during a season is adequate for crops.

Rules:
- Calculate the average rainfall.
- If the average rainfall is greater than or equal to the required level → Adequate Rainfall
- Otherwise → Inadequate Rainfall

Return the average rainfall and the rainfall status.

Real-Time Applications:
- Smart farming systems
- Irrigation planning
- Crop selection based on climate data

## Logic Explanation

1. Create a function that accepts rainfall data and required rainfall level.
2. Calculate the total rainfall using a loop.
3. Find the average rainfall.
4. Compare the average with the required rainfall level.
5. Return the average rainfall and the rainfall status.

```python
def rainfall_adequacy_checker(rainfall_data, required_level):

    total_rainfall = 0

    for rain in rainfall_data:
        total_rainfall += rain

    average_rainfall = total_rainfall / len(rainfall_data)

    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average_rainfall, status


seasonal_rainfall = [70, 65, 80, 75, 70]
required_rainfall_level = 70

avg_rainfall, rainfall_status = rainfall_adequacy_checker(
    seasonal_rainfall, required_rainfall_level
)

print("Average Rainfall:", avg_rainfall)
print("Rainfall Status:", rainfall_status)
```

## Problem 4: Social Media – Duplicate Account Detection

## Problem Statement

Create a function to detect whether duplicate usernames exist in a given list.

Rules:
- If duplicates are found → return "Duplicate accounts detected"
- Otherwise → return "All accounts are unique"

Real-Time Applications:
- Social media platform user validation
- Preventing fake or multiple account creation
- Database data cleaning

## Logic Explanation

1. Create a function that accepts a list of usernames.
2. Convert the list into a set.
3. Compare the length of the list with the length of the set.
4. If both lengths are different → duplicates exist.
5. Return the appropriate message.

```python
def detect_duplicate_accounts(usernames):

    return (
        "Duplicate accounts detected"
        if len(usernames) != len(set(usernames))
        else "All accounts are unique"
    )


users = ["dev_01", "sanya_02", "riya_03"]

print(f"Status: {detect_duplicate_accounts(users)}")
```

## Problem 6: Agriculture – Premium Crop Price Filter

## Problem Statement

Create a function to filter premium crops whose price is greater than a given threshold.

Rules:
- Accept a dictionary containing crop names and their prices.
- Accept a minimum price (threshold).
- Return the list of crops whose price is greater than the threshold.

Real-Time Applications:
- Agricultural market analysis
- Identifying high-value crops
- Farmer profit planning systems

## Logic Explanation

1. Create a function that accepts crop price data and a threshold value.
2. Initialize an empty list to store premium crops.
3. Traverse through the dictionary using a loop.
4. Check if the crop price is greater than the threshold.
5. If yes, add the crop name to the result list.
6. Return the final list.

```python
def filter_premium_crops(crop_prices, threshold):

    premium_crops = []

    for crop, price in crop_prices.items():
        if price > threshold:
            premium_crops.append(crop)

    return premium_crops


crop_data = {
    "Wheat": 2100,
    "Rice": 1800,
    "Maize": 2500,
    "Barley": 1600
}

threshold_price = 2000

result = filter_premium_crops(crop_data, threshold_price)

print("Premium Crops:", result)
```

## Problem 8: Banking – Daily Transaction Limit Checker

## Problem Statement

Create a function to check whether a user's daily transaction stays within the allowed limit.

Rules:
- Accept a list of transaction amounts for the day.
- Accept the daily transaction limit.
- Calculate the total transaction amount.
- If the total amount is less than or equal to the limit → "Within Limit"
- Otherwise → "Limit Exceeded"

Real-Time Applications:
- Banking fraud detection systems
- UPI transaction monitoring
- Account safety and spending control

## Logic Explanation

1. Create a function that accepts the transaction list and the daily limit.
2. Calculate the total transaction amount using sum().
3. Compare the total with the given limit.
4. If total ≤ limit → return "Within Limit".
5. Otherwise → return "Limit Exceeded".

```python
def check_transaction_limit(transactions, daily_limit):

    total_amount = sum(transactions)

    if total_amount <= daily_limit:
        return "Within Limit"
    else:
        return "Limit Exceeded"


transaction_list = [1200, 800, 500, 700]
limit = 3000

status = check_transaction_limit(transaction_list, limit)

print("Transactions:", transaction_list)
print("Daily Limit:", limit)
print("Total Spent:", sum(transaction_list))
print("Status:", status)
```

## Problem 9: E-Learning – Student Attendance Eligibility System

## Problem Statement

Create a function to check whether a student is eligible for an exam based on attendance.

Rules:
- A student must have at least 75% attendance to be eligible.
- Accept total number of classes and attended classes.
- Calculate the attendance percentage.
- Return:
  - "Eligible for exam"
  - "Not eligible for exam"

Real-Time Applications:
- Learning Management Systems (LMS)
- Automated academic eligibility checks
- Student performance tracking

## Logic Explanation

1. Create a function that accepts total classes and attended classes.
2. Calculate attendance percentage using:
   attendance = (attended / total) * 100
3. Check if attendance is greater than or equal to 75.
4. Return the eligibility status along with attendance percentage.

```python
def check_exam_eligibility(total_classes, attended_classes):

    attendance_percentage = (attended_classes / total_classes) * 100

    if attendance_percentage >= 75:
        return attendance_percentage, "Eligible for exam"
    else:
        return attendance_percentage, "Not eligible for exam"


total = 100
attended = 82

attendance, status = check_exam_eligibility(total, attended)

print("Total Classes:", total)
print("Attended Classes:", attended)
print("Attendance Percentage:", attendance)
print("Status:", status)
```

## Problem 10: Smart Electricity Bill Analyzer

## Problem Statement

Create a function to calculate the electricity bill based on unit consumption.

Rules:
- First 100 units → ₹5 per unit
- Next 100 units → ₹7 per unit
- Above 200 units → ₹10 per unit

Return the total bill amount.

Real-Time Applications:
- Smart meter systems
- Household electricity billing
- Energy consumption monitoring tools

## Logic Explanation

1. Create a function that accepts the number of units consumed.
2. If units ≤ 100 → bill = units × 5
3. If units ≤ 200 → bill = (100 × 5) + (remaining units × 7)
4. If units > 200 → bill = (100 × 5) + (100 × 7) + (remaining units × 10)
5. Return the total bill amount.

```python
def calculate_electricity_bill(units):

    if units <= 100:
        bill = units * 5

    elif units <= 200:
        bill = (100 * 5) + (units - 100) * 7

    else:
        bill = (100 * 5) + (100 * 7) + (units - 200) * 10

    return bill


units_consumed = 250

total_bill = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Electricity Bill: ₹", total_bill)
```
